## Measures of central tendency: 
It help you find the middle, or the average, or the centre of a dataset. The 3 most common measures of central tendency are the mode, median, and mean.
- Mode: the most frequent value.
- Median: the middle number in an ordered dataset.
- Mean: the sum of all values divided by the total number of values.


Eg : Dataset : {1,1,2,2,3,3,4,5,5,6,100}
- Mean : (32+100)/11   ==>  132/11    ==> 12
- Median : 

      - sort the dataset and count n=11
      - Since, 11 is odd no. ,
      (n+1)/2  ==> 12 / 2   ==> 6 
      - The 6'th number is the median: 3
      
      - if n would have been even e.i. n=12 and dataset={1,1,2,2,3,3,4,5,5,6,100,120}
      - n / 2	= 12 / 2 =6
      (n /2) + 1 = 7
      - The 6'th number is 3, the 7'th number is 4
      - Take the average of the two middle numbers: (3+4)/2 = 3.5
- Mode : 
             
      - Sort the numbers in ascending order : 1,1,2,2,3,3,4,5,5,6,100
      - Count how many times each number appears ==> 1 : 2, 2 : 2, 3 : 2, 4 : 1, 5 : 2, 6 : 1, 100 : 1
      - Check which numbers appear most frequently : 1, 2, 3, 5 appear most frequently (2 times), these are the modes
      


- Outlier :  An extreme value in a set of data which is much higher or lower than the other numbers. 
- Mean: Sensitive to outliers, gets pulled in their direction.
- Median: Less affected by outliers, reflects the "middle" value.
- Mode: Least affected by outliers, shows the most frequent value.

### Percentile And Quantile : 
* A percentile is a value below which a certain percentage of observation lie.
* In statistics, a k-th percentile, also known as percentile score or centile, is a score below which a given percentage k of scores in its frequency distribution falls or a score at or below which a given percentage falls.

Eg: 2,2,3,4,5,5,5,6,7,8,8,8,9,9,10,11,11,12

#### Five No Summary :
A summary consists of five values: the most extreme values in the data set (the maximum and minimum values), the lower and upper quartiles, and the median. These values are presented together and ordered from lowest to highest: 
1. minimum value 
2. Q1 (the first quartile, or the 25% mark)
3. median value (Q2) 
4. Q3 (the third quartile, or the 75% mark)
5. maximum value.

##### Quartiles are a type of percentile. A percentile is a value with a certain percentage of the data falling below it. In general terms, k% of the data falls below the kth percentile.

1. The first quartile (Q1, or the lowest quartile) is the 25th percentile, meaning that 25% of the data falls below the first quartile.
2. The second quartile (Q2, or the median) is the 50th percentile, meaning that 50% of the data falls below the second quartile.
3. The third quartile (Q3, or the upper quartile) is the 75th percentile, meaning that 75% of the data falls below the third quartile.

####  The interquartile range (IQR) : 
* It may also be called the midspread, middle 50%, fourth spread, or H‑spread. It is defined as the difference between the 75th and 25th percentiles of the data.
* To calculate the IQR, the data set is divided into quartiles, or four rank-ordered even parts via linear interpolation.
* These quartiles are denoted by Q1 (also called the lower quartile), Q2 (the median), and Q3 (also called the upper quartile). The lower quartile corresponds with the 25th percentile and the upper quartile corresponds with the 75th percentile, so IQR = Q3 −  Q1 .
* The IQR is used to build box plots, simple graphical representations of a probability distribution.
* The IQR is used in businesses as a marker for their income rates.
* The IQR can be used to identify outliers . The IQR also may indicate the skewness( the degree of asymmetry observed in a probability distribution) of the dataset.

### Removing the outliers : 
* to remove the outliers we need to define the LOWER FENCE AND UPPER FENCE
* Lower Fence = Q1 - 1.5(IQR)
* Upper Fence = Q3 + 1.5(IQR)
* IQR : InterQuantile Range = Q3 - Q1

Eg : 1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9,27   ; Q3= 75% ,   Q1=25%

Q1= (25/100) * 20 =5(index) = 3 
Q3= (75/100) * 20 =15(index) = 7

IQR = 7-3 = 4

Lower Fence = Q1 - 1.5(IQR) = 3 - 1.5(4) = -3

Upper Fence = Q3 + 1.5(IQR) = 7 - 1.5(4) = 13

[LOWER FENCE       <-->     UPPER FENCE]   ==  [-3    <----->   13]
